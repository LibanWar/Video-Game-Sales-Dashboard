```python
import numpy as np
import pandas as pd
```

```python
## IMPORT DATASET
df = pd.read_csv("vgsales.csv")
df.head(100)
```

```python
## CHECKING FOR MISSING VALUES
missing_values = df.isnull().sum()
missing_values
```

```python
## Converting Year column into integer
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

## printing data types after conversion
print("\nData types after conversion:\n", df.dtypes)
```

```python
## Dropping rows with missing 'Year' or 'Global_Sales' 
df_cleaned = df.dropna(subset=['Year', 'Global_Sales'])
```

```python
##Checking for duplicates
duplicate_rows = df_cleaned.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicate_rows}")
```

```python
## Removing any duplicate rows
df_cleaned = df_cleaned.drop_duplicates()
```

```python
#SUMMARY OF THE SALES
df_cleaned[['Global_Sales', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].describe()
```
